```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tree
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
#achar a média ariméticados três scores
df["mean score"] = df.iloc[:,5:8].sum(axis=1).divide(3)
df
```

```python


df = df.drop(["math score", "reading score", "writing score"], axis=1)

#transforma o mean score em algo binário
numero_de_linhas = len(df.index)
numero_de_colunas = len(df.columns)
for i in range(numero_de_linhas):
    if df.iloc[i, 5] >= 60:
        df.iloc[i, 5] = "aprovado"
    else:
        df.iloc[i, 5] = "reprovado"
        
df = df.rename(columns={"mean score":"aprovacao"})

#atribui a cada categoria um inteiro diferente já que o método fit não suporta strings
df["gender"] = pd.factorize(df.gender)[0]
df["race/ethnicity"] = pd.factorize(df["race/ethnicity"])[0]
df["parental level of education"] = pd.factorize(df["parental level of education"])[0]
df["lunch"] = pd.factorize(df["lunch"])[0]
df["test preparation course"] = pd.factorize(df["test preparation course"])[0]
df
```

```python
x = df.drop(["aprovacao"], axis=1)
y = df["aprovacao"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# checando se há uma string
condicao = False
for i in range(numero_de_linhas):
    for j in range(numero_de_colunas-1):
        if type(x.iloc[i, j]) is str:
            condicao = True
            
if condicao == True:
    print("There's a string here")

    
model = LogisticRegression()
model.fit(x_train, y_train)
resultado = model.predict(x_test)
print(accuracy_score(resultado, y_test))
print(confusion_matrix(resultado, y_test))
```
